# Limpeza de Dados (data-cleaning)

```python
import pandas as pd
df = pd.read_csv("british_library.csv")
df
```

```python
df.info()
# Mostra muitas colunas com dados nulos que de nada servem
```

```python
# Então vamos apagar essas colunas.
df.drop(
    columns=[
        'Edition Statement', 
        'Corporate Author', 
        'Corporate Contributors', 
        'Former owner', 
        'Engraver'])
```

```python
# Se observarmos as colunas ainda existem.
df.info()
```

```python
# Para deletar realmente as colunas temos de usar inplace()
df.drop(
    columns=[
        'Edition Statement', 
        'Corporate Author', 
        'Corporate Contributors', 
        'Former owner', 
        'Engraver'], inplace=True)
```

```python
# Ao usar inplace percebemos que as colunas realmente desapareceram.
df.info()
```

```python
df.head()
```

**OBS.: Agora já com as colunas nulas eliminadas temos algumas sobre dados que são ou não realmente úteis: URL, Data de publicação em formatos diferentes, Número de Identificação...**
